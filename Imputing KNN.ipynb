```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.impute import KNNImputer
#INPUT = 'CompleteIndexes.csv'
#OUTPUT = 'KNNImputed.csv'
INPUT = 'CompleteWeeklyIndexes.csv'
OUTPUT = 'KNNWeeklyImputed.csv'
```

```python
df = pd.read_csv(INPUT, parse_dates=True, index_col=0)
df = df.loc['2000-01-01':'2015-01-01'] #Selecting training data
df.head()
```

```python
imp = KNNImputer(n_neighbors=5, weights="uniform")
```

```python
df_imputed = imp.fit_transform(df)
```

```python
df_final = pd.DataFrame(data=df_imputed, index=df.index, columns=df.columns)
```

```python
df_final.tail()
```

```python
df_final.to_csv(OUTPUT, index = True)
```
