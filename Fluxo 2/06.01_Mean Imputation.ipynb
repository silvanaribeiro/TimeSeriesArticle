## This notebook inputs the missing values using Mean Imputation

```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.impute import SimpleImputer
```

#### Choose wether the imputation should be done on the data set with daily or weekly granularity

```python
#INPUT, OUTPUT = 'CompleteIndexes.csv','MeanImputed.csv'
INPUT, OUTPUT = 'CompleteWeeklyIndexes', 'MeanWeeklyImputed'
```

```python
df = pd.read_csv(INPUT+'_training_FE.csv', parse_dates=True, index_col=0)
df_test = pd.read_csv(INPUT+'_test_FE.csv', parse_dates=True, index_col=0)
df.head()
```

```python
df.tail()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(df)
```

```python
df_imputed = imp.transform(df)
df_test_imputed = imp.transform(df_test)
```

```python
df_final = pd.DataFrame(data=df_imputed, index=df.index, columns=df.columns)
df_test_final = pd.DataFrame(data=df_test_imputed, index=df_test.index, columns=df_test.columns)
```

```python
df_final.head()
```

```python
df_final.to_csv(OUTPUT+'_training_FE.csv', index = True)
df_test_final.to_csv(OUTPUT+ '_test_FE.csv', index = True)
```
