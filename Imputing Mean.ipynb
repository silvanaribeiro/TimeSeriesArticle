```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('CompleteIndexes.csv', parse_dates=True, index_col=0)
df.head()
```

```python
df.tail()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(df)
```

```python
df_imputed = imp.transform(df)
```

```python
df_final = pd.DataFrame(data=df_imputed, index=df.index, columns=df.columns)
```

```python
print(df_imputed)
```

```python
df_final.to_csv(r'MeanImputed.csv', index = True)
```
