```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
df = pd.read_csv('CompleteIndexes.csv', parse_dates=True, index_col=0)
df.head()
```

```python
imp = IterativeImputer(max_iter=10, random_state=0)
```

```python
imp.fit(df)
```

```python
df_imputed = imp.transform(df)
```

```python
df_final = pd.DataFrame(data=df_imputed, index=df.index, columns=df.columns)
```

```python
df_final.head()
```

```python
df_final.to_csv(r'RegressionImputed.csv', index = True)
```
