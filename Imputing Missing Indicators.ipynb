```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.impute import MissingIndicator
```

```python
df = pd.read_csv('CompleteIndexes.csv', parse_dates=True, index_col=0)
df.head()
```

```python
indicator = MissingIndicator(missing_values=np.nan, features="all")
```

```python
mask_all = indicator.fit_transform(df)
```

```python
mask_all
```

```python
df_masks = pd.DataFrame(data=mask_all, index=df.index, columns=df.columns)
```

```python
df_masks.add_prefix('MI_')
```

```python
df_final = pd.concat([df, df_masks], axis=1, join='outer')
```

```python
df_final.head()
```

```python
df.head()
```

```python
df_final.to_csv(r'MIImputed.csv', index = True)
```
